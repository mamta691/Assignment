```python

sum=0
product=1
i=1
while i<=10:
    sum+=i
    product*=i
    i+=1
    print("sum of 10 natural no is:",sum)
    print("product of 10 natural numbers:",product)
```

Q4.Create a list of numbers from 1 to 100. Use for loop and while loop to clculate the cube of each number and if that numbers is divisible by 4 or 5 then append that no. in a list and print that list?

```python
num=[]
cube_divisible=[]

for i in range(1, 101):
    num.append(i)
    cube=i**3
    if cube%4==0 or cube%5==0:
        cube_divisible.append(cube)
    print("The list of cube that is divisible by 4 or 5:",cube_divisible)
```

Q5.Write a program to print count vowels in a given string"i want to become a data scientist"?

```python
String="i want to become a data scientist"

count=0

vowels=["a","e","i","o","u"]

for i in range(len(String)):
    if String[i] in vowels:
        count=count+1
        print("No. of vowels in the string is:",count)
```
